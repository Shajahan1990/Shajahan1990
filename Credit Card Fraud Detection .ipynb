#### Oversampling

* We use smote that means synthetic minority oversampling technique. 
* its one if the most common method for oversampling to solve data imbalancing
* smote used to balance class distribution by randomly increasing minority class example by replicating them
* smote synthesize new minority instances between exising minority instances

```python
x = df.drop(['Class'], axis = 1)
y = df['Class']
```

```python
x.shape, y.shape
```

```python
from imblearn.over_sampling import SMOTE
x_res, y_res = SMOTE().fit_resample(x,y)
```

```python
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_res,y_res, test_size = 0.2, random_state = 42)
```
